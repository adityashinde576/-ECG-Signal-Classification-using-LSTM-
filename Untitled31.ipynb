# Import Libraries


```python

import os
import zipfile
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

# Check GPU


```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

# Load Dataset


```python
dataset_path = "heartbeat.zip"
if not os.path.exists("data/mitbih_train.csv"):
    os.system("kaggle datasets download -d shayanfazeli/heartbeat -f mitbih_train.csv -p data/")
    os.system("kaggle datasets download -d shayanfazeli/heartbeat -f mitbih_test.csv -p data/")
with zipfile.ZipFile("data/mitbih_train.csv.zip", 'r') as zip_ref:
    zip_ref.extractall("data/")
with zipfile.ZipFile("data/mitbih_test.csv.zip", 'r') as zip_ref:
    zip_ref.extractall("data/")
```

```python
train_df = pd.read_csv("data/mitbih_train.csv", header=None)
test_df  = pd.read_csv("data/mitbih_test.csv", header=None)

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
```

# Preprocessing


```python
X_train = train_df.iloc[:, :-1].values
y_train = train_df.iloc[:, -1].values
X_test  = test_df.iloc[:, :-1].values
y_test  = test_df.iloc[:, -1].values
```

# Standardize features


```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

# Reshape for LSTM: (samples, time_steps, features)

```python
X_train = X_train.reshape(-1, 187, 1)
X_test  = X_test.reshape(-1, 187, 1)
```

# Encode labels


```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test  = le.transform(y_test)
```

# Convert to Torch tensors

```python
train_dataset = TensorDataset(torch.Tensor(X_train), torch.LongTensor(y_train))
test_dataset  = TensorDataset(torch.Tensor(X_test), torch.LongTensor(y_test))

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader  = DataLoader(test_dataset, batch_size=128, shuffle=False)
```

 # Define LSTM Model

```python
class ECG_LSTM(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_classes=5):
        super(ECG_LSTM, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc   = nn.Linear(hidden_size, num_classes)
    
    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]  
        out = self.fc(out)
        return out

model = ECG_LSTM().to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

# Training Loop


```python
epochs = 20
for epoch in range(epochs):
    model.train()
    running_loss = 0
    for X_batch, y_batch in train_loader:
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)
        
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader):.4f}")
```

# Evaluation


```python
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for X_batch, y_batch in test_loader:
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)
        
        outputs = model(X_batch)
        _, predicted = torch.max(outputs.data, 1)
        total += y_batch.size(0)
        correct += (predicted == y_batch).sum().item()

print(f"Test Accuracy: {correct/total:.4f}")
```

# Save Model


```python
torch.save(model.state_dict(), "ecg_lstm_model_gpu.pth")
```

# Visualize Sample ECG Signals


```python
import matplotlib.pyplot as plt
num_samples = 5
for i in range(num_samples):
    signal = X_test[i].reshape(-1)  
    label = le.inverse_transform([y_test[i]])[0]
    
    plt.figure(figsize=(10, 3))
    plt.plot(signal)
    plt.title(f"ECG Signal #{i+1} - True Label: {label}")
    plt.xlabel("Time Steps")
    plt.ylabel("Amplitude")
    plt.show()
```

# Model Evaluation and Label Inversion


```python
model.eval()
predictions = []
true_labels = []
with torch.no_grad():
    for X_batch, y_batch in test_loader:
        X_batch = X_batch.to(device)
        outputs = model(X_batch)
        _, predicted = torch.max(outputs.data, 1)
        predictions.extend(predicted.cpu().numpy())
        true_labels.extend(y_batch.cpu().numpy())
pred_labels = le.inverse_transform(predictions)
true_labels = le.inverse_transform(true_labels)
```

# Visualizing Sample ECG Predictions

```python

num_samples = 5
for i in range(num_samples):
    signal = X_test[i].reshape(-1)
    true_label = true_labels[i]
    pred_label = pred_labels[i]
    
    plt.figure(figsize=(10, 3))
    plt.plot(signal)
    plt.title(f"ECG Signal #{i+1} - True: {true_label}, Predicted: {pred_label}")
    plt.xlabel("Time Steps")
    plt.ylabel("Amplitude")
    plt.show()
```

# Confusion Matrix Visualization

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.heatmap(pd.crosstab(np.array(true_labels), np.array(pred_labels)), annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```
